**1.Prime Numbers**

```python
num = int(input("Enter number: "))

for i in range(2, num):
    if num % i == 0:
        print("Not Prime")
        break
else:
    print("Prime")
```

**2.Product of Random Numbers**

```python
import random
a=random.randint(1,10)
b=random.randint(1,10)
print("Numbers are:", a, "and", b)
num=int(input('Enter the answer:'))
if num==a*b:
  print('correct')
else:
  print('wrong')
```

**3.Squares of Even/Odd Numbers**

```python
## program to print square of even/odd numbers fromm 100 to 201

print('Squares of Even numbers')
for i in range(100,201):
    if i%2==0:
      print('Square of',i,'is',i*i)

print('Squares of Odd numbers')
for i in range(100,201):
    if i%2!=0:
      print('Square of',i,'is',i*i)

```

**Exercise 4**:Problem Statement (Question): Hospital Billing System


```python
room_charge_per_day=2000
doctor_fee=1500
lab_test_cost=300

days=4
lab_test=3
medicine_charges=2400

##calculating charges
room_charges=room_charge_per_day*days
lab_charges=lab_test_cost*lab_test

##total bill before discount
total_bill=room_charges+lab_charges+medicine_charges+doctor_fee

##applying discounts
if total_bill>10000:
   discount=total_bill*0.10
else:
   discount=0

# Calculate final bill after discount
final_bill = total_bill - discount

##displaying the bills
print("Final Bill =", final_bill)
```
