**1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.**

```python
##Create a list of 5 integers
num=[10,20,30,40,50]
print('Original List is:',num)

##append()
num.append(60)
print('After append(60) list is:',num)

##extend
num.extend([70,80])
print('After extend ([60,70]) list is:',num)

##insert()
num.insert(2,25)
print('After inserting 25 list is:',num)

##remove()
num.remove(30)
print('After removing 30 list is:',num)

##pop()
removed_item = num.pop()
print("After pop():", num)
print("Popped Item:", removed_item)

##index()
print('Index of 40 is:',num.index(40))

##count()
print('Count of 50 is:',num.count(50))

##sort
num.sort()
print('Sorted list is',num)

##Reverse()
num.reverse()
print('Reversed list is:',num)

##clear()
num.clear()
print("After clear():", num)
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows **bold text**


```python
import pandas as pd
import numpy as np

# Create DataFrame with 10 rows
data = {
    'Name': ['Asha', 'Rahul', 'Priya', 'Kiran', 'Sneha',
             'Arjun', 'Meena', 'Ravi', 'Pooja', 'Rahul'],
    'Age': [21, 22, 20, 23, 21, 24, 22, 23, 20, 22],
    'City': ['Bangalore', 'Hyderabad', 'Chennai', 'Mumbai', 'Delhi',
             'Pune', 'Kolkata', 'Jaipur', 'Lucknow', 'Hyderabad'],
    'Salary': [30000, 40000, np.nan, 45000, 38000,
               50000, np.nan, 42000, 36000, 40000]
}

df = pd.DataFrame(data)

# 1. Display info
print("DataFrame Info:")
df.info()

# 2. Display description
print("\nDataFrame Description:")
print(df.describe())

# 3. Select Name and City columns
print("\nName and City Columns:")
print(df[['Name', 'City']])

# 4. Drop City column
df = df.drop('City', axis=1)
print("\nAfter Dropping City Column:")
print(df)

# 5. Fill missing Salary values with mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\nAfter Filling Missing Salary Values:")
print(df)

# 6. Remove duplicate rows
df = df.drop_duplicates()
print("\nAfter Removing Duplicate Rows:")
print(df)
```
