# Problem Statement : Hospital Patient Data Analysis

```python
import numpy as np
import pandas as pd
```

**1.Load the patient dataset and show summary with info()**

```python
df=pd.read_csv('Patient_Data.csv')
```

```python
df.info()
```

 **2. Select only the columns relevant for billing**


```python
billing_columns = df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
print("\nBilling Related Columns:")
print(billing_columns.head())
```


**3. Drop administrative columns**

```python
df = df.drop(columns=['ReceptionistID', 'CheckInTime'])
```

 **4. Use groupby to find total bill amount per department**

```python
department_bill = df.groupby('Department')['BillAmount'].sum()
print("\nTotal Bill Amount Per Department:")
print(department_bill)
```

**5. Remove duplicate patient records based on PatientID**

```python
df = df.drop_duplicates(subset='PatientID')
```

```python
df.duplicated().sum()
```

**6. Fill missing BillAmount values with the mean bill amoun**t

```python
mean_bill = df['BillAmount'].mean()
df['BillAmount'] = df['BillAmount'].fillna(mean_bill)
```

```python
df.isnull().sum()
```

**7. Merge the billing dataset with patient dataset on PatientID**

```python
billing_df = pd.read_csv("Billing_Data.csv")
merged_df = pd.merge(df, billing_df, on='PatientID', how='inner')
```

```python
print("\nMerged Dataset:")
print(merged_df.head())
```

**8. Concatenate an additional DataFrame containing new patients (row-wise)**

```python
new_patients_df = pd.DataFrame({
    'PatientID': [201, 202],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Kumar', 'Dr. Rao'],
    'BillAmount': [4500, 6000]
})

combined_df = pd.concat([merged_df, new_patients_df], axis=0, ignore_index=True)

print("After Adding New Patients:")
print(combined_df)
```

**9. Concatenate new billing category columns like # ['InsuranceCovered', 'FinalAmount'] (column-wise)**

```python
billing_categories_df = pd.DataFrame({
    'InsuranceCovered': [1000] * len(combined_df),
    'FinalAmount': [4000] * len(combined_df)
})

final_dataset = pd.concat([combined_df, billing_categories_df], axis=1)

print("Final Dataset:")
print(final_dataset)
```
