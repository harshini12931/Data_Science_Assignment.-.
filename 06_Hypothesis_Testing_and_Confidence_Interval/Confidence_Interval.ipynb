# ESTIMATION AND CONFIDENCE INTERVALS

**a. Build 99% Confidence Interval Using Sample Standard Deviation**

```python
##importing required libraries
import numpy as np
from scipy import stats
```

```python
##storing sample data

data = [1.13, 1.55, 1.43, 0.92, 1.25,
        1.36, 1.32, 0.85, 1.07, 1.48,
        1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
##Calculating Sample Mean and Standard Deviation

n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)

print("Sample Mean =", mean)
print("Sample Standard Deviation =", std)
```

```python
##Calculating 99% Confidence Interval Using t-Distribution

confidence = 0.99

t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

margin_error = t_critical * (std / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

print("99% Confidence Interval =", (lower, upper))
```

*Explanation : Since the sample size is small (n = 15) and the population standard deviation is unknown, the t-distribution is used. The 99% confidence interval provides a range within which the true mean durability of print-heads is expected to lie with 99% confidence*.

**b. Build 99% Confidence Interval Using Known Population Standard Deviation**

```python
##Define Population Standard Deviation
sigma=0.2
```

```python
##Calculating 99% Confidence Interval Using z-Distribution

z_critical = stats.norm.ppf(0.995)

margin_error = z_critical * (sigma / np.sqrt(n))

lower = mean - margin_error
upper = mean + margin_error

print("99% Confidence Interval =", (lower, upper))
```

*Explanation :  Since the population standard deviation is known (0.2 million characters), the z-distribution is used. The resulting 99% confidence interval estimates the true mean durability of the print-heads with a confidence level of 99%.*
