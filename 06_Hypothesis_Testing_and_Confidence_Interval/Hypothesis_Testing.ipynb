# HYPOTHESIS TESTING

```python
import math
from scipy import stats
import numpy as np
import pandas as pd
```

**1.State the Hypotheses statement:**

```python
# Null Hypothesis (H0): Weekly operating cost is equal to the model cost
# Alternative Hypothesis (H1): Weekly operating cost is higher than the model cost
print("H0: μ = 4000")
print("H1: μ > 4000")
```

**2. Calculate the Test Statistic:**

```python
# Given values
x_bar = 3050
mu = 1000 + (5 * 600)
sigma = 5 * 25
n = 25

# Calculate z statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))

print("Theoretical Mean Cost =", mu)
print("Z Statistic =", z)
```

**3. Determine the Probability and compare:**

```python
alpha = 0.05

# Critical value for right-tailed test
critical_value = stats.norm.ppf(1 - alpha)

print("Critical Value =", critical_value)
print("Calculated Z Value =", z)
```

**4. Make a Decision:**

```python
if z > critical_value:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

**5. Conclusion**

```python
if z > critical_value:
    print("There is strong evidence that weekly operating costs are higher than the model suggests.")
else:
    print("There is no evidence that weekly operating costs are higher than the model suggests.")
```

> **Conclusion:**
> *Based on the hypothesis test, there is not enough evidence to support the restaurant owners' claim that weekly operating costs have increased beyond what the existing cost model predicts. The sample average weekly cost was actually lower than the theoretical cost calculated using the model. Therefore, the current data does not indicate that the operating cost model needs to be revised at this time.*
